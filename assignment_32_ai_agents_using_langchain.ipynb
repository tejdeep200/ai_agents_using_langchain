# Task 1: Understanding Tools (Conceptual)

### 1. What is a tool in an AI Agent?
A tool in an AI agent is an external function or capability that the agent can use to perform specific tasks beyond its built-in knowledge. Examples include calculators, APIs, databases, or search engines.

---

### 2. Why do agents need tools?
Agents need tools because:
- They extend the agent’s capabilities beyond text generation.
- They allow real-time data access (e.g., web search).
- They enable accurate computations and actions.
- They help interact with external systems (APIs, databases, etc.).

---

### 3. Difference between a chatbot and an agent

| Feature        | Chatbot                          | AI Agent                              |
|---------------|----------------------------------|----------------------------------------|
| Functionality  | Responds to user input           | Performs tasks using tools             |
| Capabilities   | Limited to conversation          | Can take actions and make decisions    |
| Tool Usage     | Usually no tools                 | Uses multiple tools (APIs, DBs, etc.)  |
| Autonomy       | Low                              | Higher autonomy                        |

````python
from langchain.tools import tool
from datetime import datetime

@tool
def company_policy_lookup(query: str) -> str:
    policies = {
        "leave": "Employees get 20 days of paid leave per year.",
        "remote": "Remote work allowed 3 days per week.",
        "dress": "Business casual dress code."
    }
    for key in policies:
        if key in query.lower():
            return policies[key]
    return "Policy not found."

@tool
def employee_db_lookup(name: str) -> str:
    db = {
        "john": "John is a Software Engineer.",
        "sarah": "Sarah is a Data Analyst.",
        "raj": "Raj is a Project Manager."
    }
    return db.get(name.lower(), "Employee not found.")

@tool
def get_current_time(_: str) -> str:
    return f"{datetime.now()}"

toolkit = [company_policy_lookup, employee_db_lookup, get_current_time]

for tool in toolkit:
    print(tool.name)
````

#task 5

````python
from langchain.chat_models import ChatOpenAI
from langchain.agents import initialize_agent, AgentType

llm = ChatOpenAI(temperature=0)

agent = initialize_agent(
    tools=toolkit,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

response = agent.run("What is the leave policy?")
print(response)
````

# Task 7: ReAct Agent Overview

## What is ReAct (Reason + Act)?
ReAct is a framework where an AI agent combines reasoning and actions.
The agent:
- Thinks about the problem (Reason)
- Decides what action to take (Act)
- Observes the result
- Repeats until final answer

## Why ReAct agents are powerful
- Enables step-by-step reasoning
- Can dynamically use tools
- Handles complex multi-step problems
- Improves accuracy by verifying with tools

````python
from langchain.tools import tool
from langchain.chat_models import ChatOpenAI
from langchain.agents import initialize_agent, AgentType
from langchain.utilities import WikipediaAPIWrapper
from langchain.tools import Tool
from langchain.agents import load_tools

wiki = WikipediaAPIWrapper()
wikipedia_tool = Tool(
    name="Wikipedia",
    func=wiki.run,
    description="Search Wikipedia"
)

calculator = load_tools(["llm-math"])[0]

@tool
def get_current_time(_: str) -> str:
    from datetime import datetime
    return str(datetime.now())

tools = [wikipedia_tool, calculator, get_current_time]

llm = ChatOpenAI(temperature=0)

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

response = agent.run("Who is Albert Einstein?")
print(response)
````

````python
from langchain.tools import tool
from langchain.utilities import WikipediaAPIWrapper
from langchain.tools import Tool
from langchain.agents import initialize_agent, AgentType, load_tools
from langchain.chat_models import ChatOpenAI

wiki = WikipediaAPIWrapper()
wikipedia_tool = Tool(name="Wikipedia", func=wiki.run, description="Search Wikipedia")

calculator = load_tools(["llm-math"])[0]

@tool
def simple_search(query: str) -> str:
    data = {
        "ai": "AI stands for Artificial Intelligence.",
        "python": "Python is a programming language."
    }
    return data.get(query.lower(), "Not found")

tools = [wikipedia_tool, calculator, simple_search]

llm = ChatOpenAI(temperature=0)

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

print(agent.run("What is AI?"))
print(agent.run("Who is Narendra Modi?"))
print(agent.run("What is 100 * 5?"))
````

# Task 11: Observations & Insights

## Benefits of tool-augmented agents
- Access real-time and external data
- Perform accurate computations
- Solve complex multi-step problems
- More flexible than simple LLMs

## Challenges with agents
- Slower due to multiple steps
- Tool errors can affect output
- Requires proper tool design
- Higher cost (API calls)

## Difference between chains and agents
- Chains: Fixed sequence of steps
- Agents: Dynamic decision-making system

## When to use agents over RAG
- When tasks require actions (API calls, tools)
- When multi-step reasoning is needed
- When dynamic decision-making is required
- RAG is better for document-based Q&A only
